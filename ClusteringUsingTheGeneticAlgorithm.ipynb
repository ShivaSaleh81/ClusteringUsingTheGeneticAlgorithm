```python
import numpy as np

def initialize_population(population_size, num_data_points, max_clusters, dataset):
    population = []
    for _ in range(population_size):
        chromosome = np.random.randint(1, max_clusters + 1, size=num_data_points)
        num_clusters = np.max(chromosome)
        chromosome = np.append(chromosome, num_clusters)
        population.append(chromosome)

    return population

population_size = 10
num_data_points = 150
max_clusters = 3

initial_population = initialize_population(population_size, num_data_points, max_clusters, data)
for i, chromosome in enumerate(initial_population):
    print(f"Chromosome {i + 1}: {chromosome}")
```

```python
import numpy as np

def calculate_fitness(chromosome, data, num_clusters):
    chromosome_array = chromosome[:-1]
    num_clusters = int(chromosome[-1])

    cluster_centers = []
    counts = []
    for i in range(num_clusters):
      cluster_centers.append(np.zeros(4))
      counts.append(0)

    for i, datapoint_cluster in enumerate(chromosome_array):
        cluster_centers[datapoint_cluster - 1] += data[i]
        counts[datapoint_cluster - 1] += 1

    for i, sum in enumerate(cluster_centers):
      sum /= counts[i]

    distances = np.zeros((data.shape[0], num_clusters))
    for i, point in enumerate(data):
        feature_distances = np.sqrt(np.sum((point - cluster_centers)**2, axis=1))
        distances[i] = feature_distances

    assigned_clusters = np.argmin(distances, axis=1) + 1
    cluster_sums = np.zeros(num_clusters)
    for cluster in range(1, num_clusters + 1):
        cluster_points = data[assigned_clusters == cluster]
        if len(cluster_points) > 0:
            feature_distances = np.sqrt(np.sum((cluster_points - cluster_centers[cluster - 1])**2, axis=1))
            cluster_sums[cluster - 1] = np.sum(feature_distances)

    total_distance = np.sum(cluster_sums)
    fitness = 1 / total_distance
    return fitness

for i, chromosome in enumerate(initial_population):
    fitness_value = calculate_fitness(chromosome, data, num_clusters=3)
    print(f"Fitness for Chromosome {i + 1}: {fitness_value}")
```

```python
def crossover(parent1, parent2):
    crossover_point = np.random.randint(1, len(parent1) - 1)
    offspring1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
    offspring2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
    return offspring1, offspring2
```

```python
def mutation(chromosome, mutation_probability=0.2):
    if np.random.rand() < mutation_probability:
        num_genes_to_invert = np.random.randint(1, len(chromosome) // 2)
        genes_to_invert = np.random.choice(len(chromosome), size=num_genes_to_invert, replace=False)
        chromosome[genes_to_invert] = np.flip(chromosome[genes_to_invert])

    return chromosome
```

```python
# Genetic Algorithm Loop
num_generations = 50
mutation_probability = 0.2

best_solution = None
best_fitness = float('-inf')

for generation in range(num_generations):

    parents_indices = np.argsort(fitness_values)[-2:]
    parent1 = initial_population[parents_indices[0]]
    parent2 = initial_population[parents_indices[1]]

    # Perform crossover
    offspring1, offspring2 = crossover(parent1, parent2)

    # Perform mutation
    if np.random.rand() < mutation_probability:
        offspring1 = mutation(offspring1)
    if np.random.rand() < mutation_probability:
        offspring2 = mutation(offspring2)

    least_fit_indices = np.argsort(fitness_values)[:2]
    initial_population[least_fit_indices[0]] = offspring1
    initial_population[least_fit_indices[1]] = offspring2

    fitness_values = [calculate_fitness(chromosome, data, num_clusters=3) for chromosome in initial_population]
    current_best_fitness = max(fitness_values)
    print(f"Generation {generation + 1}, Best Fitness: {current_best_fitness}")

    if current_best_fitness > best_fitness:
        best_solution = initial_population[np.argmax(fitness_values)]
        best_fitness = current_best_fitness

print(f"Best Solution: {best_solution}")
print(f"Best Fitness: {best_fitness}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

best_cluster_assignments = best_solution[:-1]
for cluster in range(1, int(np.max(best_cluster_assignments)) + 1):
    cluster_points = data[best_cluster_assignments == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster}')

plt.title('Final Clustering Solution')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris_df = pd.DataFrame(data, columns=feature_names)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(iris_df)

iris_df['Cluster'] = kmeans.labels_

plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_df['Cluster'], cmap='viridis')
plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
